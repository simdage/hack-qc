```python
import warnings
warnings.filterwarnings("ignore")

import sys
import os
from pathlib import Path
main_folder=str(Path.cwd().parent) 
sys.path.append(main_folder)  
# data_folder = f'{main_folder}/data'

import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

from data_processing.data_cleaning import get_df_cas_1, clean_complete_series
import numpy as np

%matplotlib inline  
import plotly.express as px
```

```python
def read_csv(path):
    df = pd.read_csv(path, skiprows=2) #.iloc[:, 1:]
    return df

def get_df_cas_1():
    df_ba_07_15_b = read_csv("/Users/zarreennaowalreza/Documents/iuvo-ai/hack-qc/data/beauharnois_aval_2007_2015_brutes.csv")
    df_ba_15_22_b = read_csv("/Users/zarreennaowalreza/Documents/iuvo-ai/hack-qc/data/beauharnois_aval_2015_2022_brutes.csv")

    df_ba_07_15_v = read_csv("/Users/zarreennaowalreza/Documents/iuvo-ai/hack-qc/data/beauharnois_aval_2007_2015_validees.csv")
    df_ba_15_22_v = read_csv("/Users/zarreennaowalreza/Documents/iuvo-ai/hack-qc/data/beauharnois_aval_2015_2022_validees.csv")

    df_ba_v = pd.concat([df_ba_07_15_v, df_ba_15_22_v])
    df_ba_b = pd.concat([df_ba_07_15_b, df_ba_15_22_b])
    
#     display(df_ba_b.head())
#     display(df_ba_v.head())

    df_ba_v = df_ba_v.rename(columns={"Valeur": "Validee_aval"})
    df_ba_b = df_ba_b.rename(columns={"Valeur": "Brutte_aval"})
    
#     display(df_ba_b.head())
#     display(df_ba_v.head())

    del df_ba_15_22_v, df_ba_15_22_b, df_ba_07_15_v, df_ba_07_15_b
    df_aval = pd.merge(df_ba_b, df_ba_v, on="Date")

    df_qu_07_15_b = read_csv("/Users/zarreennaowalreza/Documents/iuvo-ai/hack-qc/data/quai_de_beauharnois_2007_2015_brutes.csv")
    df_qu_07_15_v = read_csv("/Users/zarreennaowalreza/Documents/iuvo-ai/hack-qc/data/quai_de_beauharnois_2007_2015_validees.csv")

    df_qu_15_22_b = read_csv("/Users/zarreennaowalreza/Documents/iuvo-ai/hack-qc/data/quai_de_beauharnois_2015_2022_brutes.csv")
    df_qu_15_22_v = read_csv("/Users/zarreennaowalreza/Documents/iuvo-ai/hack-qc/data/quai_de_beauharnois_2015_2022_validees.csv")


    df_qu_v = pd.concat([df_qu_07_15_v, df_qu_15_22_v])
    df_qu_b = pd.concat([df_qu_07_15_b, df_qu_15_22_b])

    del df_qu_15_22_v, df_qu_15_22_b, df_qu_07_15_v, df_qu_07_15_b

    df_qu_v = df_qu_v.rename(columns={"Valeur": "Validee_quai"})
    df_qu_b = df_qu_b.rename(columns={"Valeur": "Brutte_quai"})

    df_quai = pd.merge(df_qu_b, df_qu_v, on="Date")

    del df_ba_b, df_ba_v, df_qu_v, df_qu_b

    df = pd.merge(df_aval, df_quai, on="Date")

    del df_aval, df_quai

    df["Error_aval"] = df["Brutte_aval"] != df["Validee_aval"]
    df["Diff_aval"] = df["Brutte_aval"] - df["Validee_aval"]
    df["Error_quai"] = df["Brutte_quai"] != df["Validee_quai"]
    df["Diff_quai"] = df["Brutte_quai"] - df["Validee_quai"]

    df["Date"] = pd.to_datetime(df["Date"])

    df["Year"] = df["Date"].dt.year
    df["Month"] = df["Date"].dt.month
    df["Day"] = df["Date"].dt.day

    df["Hour"] = df["Date"].dt.hour
    print("df ready")
    return df

```

```python
df = get_df_cas_1()
print(df.shape)
df.head()
```

```python
# df = get_df_cas_1()
aval_clean, aval_index = clean_complete_series(df.copy(), 20, 24, "Brutte_aval")
quai_clean, quai_index = clean_complete_series(df.copy(), 20, 24, "Brutte_quai")

df_series = pd.concat([aval_clean, quai_clean], axis = 1)

df1 = df_series.fillna(0)

anomaly_inputs = ['Brutte_aval', 'Brutte_quai']
model_IF = IsolationForest(contamination=0.1, random_state=42)
model_IF.fit(df1[anomaly_inputs])
df1['anomaly_scores'] = model_IF.decision_function(df1[anomaly_inputs])
df1['anomaly'] = model_IF.predict(df1[anomaly_inputs])
# df1[df1["anomaly_scores"] < -0.155] = np.nan

df_series = pd.merge(df, df)
# df1['Date']
```

```python
df1['Date'] = df['Date']
print(min(df1['Date']))
print(max(df1['Date']))
```

```python
df_sampled = df1.set_index('Date').resample('H').mean().reset_index()
print(df_sampled.shape)
df_sampled.head()
```

```python
df_sampled_orig = df.set_index('Date').resample('H').mean().reset_index()
print(df_sampled_orig.shape)
df_sampled_orig.head()
```

```python
train_df = df_sampled[(df_sampled['Date'] >= '2017-01-01 00:00:00') & (df_sampled['Date'] <= '2020-01-01 00:00:00')]
test_df = df_sampled[(df_sampled['Date'] > '2020-01-01 00:00:00') & (df_sampled['Date'] <= '2021-10-17 12:00:00')]

train_df = train_df.rename(columns={"Date": "ds", "Brutte_aval": "y"})
test_df = test_df.rename(columns={"Date": "ds", "Brutte_aval": "y"})

validee_test_df = df_sampled_orig[(df_sampled_orig['Date'] > '2020-01-01 00:00:00') & (df_sampled_orig['Date'] <= '2021-10-17 12:00:00')]

print(train_df.shape)
print(test_df.shape)
print(validee_test_df.shape)
```

```python
prophet_basic = Prophet()
prophet_basic.fit(train_df)
```

```python
forecast = prophet_basic.predict(test_df)
print(len(forecast))
```

```python
fig1 = prophet_basic.plot(forecast)
```

```python
fig1 = prophet_basic.plot_components(forecast)
```

```python
plt.figure(figsize=(20,6))
plt.title("Forecasted vs Validee Data")

plt.plot(forecast['ds'], forecast["yhat"], label = "Brutte_aval Forecasted Data")
plt.plot(test_df['ds'], test_df["y"], label = "Brutte_aval Test Data")
plt.plot(validee_test_df['Date'], validee_test_df["Validee_aval"], label = "Validee_aval Data")
plt.legend()
plt.show()
```
